```python
import pandas as pd

# Load the dataset
df = pd.read_csv('../../csv/Beras.csv')

# Set the index to 'Tanggal' and select only the 'Harga Beras' column
df = df.set_index('Tanggal')
df = df[['Harga Beras']]

# Display the first few rows of the modified dataframe
df.head()

from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense
import numpy as np

# Data Preprocessing
scaler = MinMaxScaler(feature_range=(0, 1))
df_scaled = scaler.fit_transform(df)

# Function to create dataset for LSTM
def create_dataset(dataset, look_back=1):
    dataX, dataY = [], []
    for i in range(len(dataset) - look_back):
        a = dataset[i:(i + look_back), 0]
        dataX.append(a)
        dataY.append(dataset[i + look_back, 0])
    return np.array(dataX), np.array(dataY)

# Prepare input for LSTM
look_back = 1
X, y = create_dataset(df_scaled, look_back)

# Reshape input to be [samples, time steps, features]
X = np.reshape(X, (X.shape[0], 1, X.shape[1]))

# Splitting dataset into train and test sets
train_size = int(len(X) * 0.8)
test_size = len(X) - train_size
trainX, testX = X[0:train_size], X[train_size:len(X)]
trainY, testY = y[0:train_size], y[train_size:len(y)]

# LSTM Network
model = Sequential()
model.add(LSTM(4, input_shape=(1, look_back)))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(trainX, trainY, epochs=100, batch_size=1, verbose=2)

# Making predictions
trainPredict = model.predict(trainX)
testPredict = model.predict(testX)

# Invert predictions
trainPredict = scaler.inverse_transform(trainPredict)
trainY = scaler.inverse_transform([trainY])
testPredict = scaler.inverse_transform(testPredict)
testY = scaler.inverse_transform([testY])

# Forecast 90 days ahead
input_seq = df_scaled[-look_back:]
predictions = []
for _ in range(90):
    pred = model.predict(input_seq.reshape(1, 1, look_back))
    predictions.append(pred[0, 0])
    input_seq = np.append(input_seq[1:], pred)

predictions = scaler.inverse_transform(np.array(predictions).reshape(-1, 1))

import pandas as pd
import matplotlib.pyplot as plt

# Ensure datetime conversion is correct
df.index = pd.to_datetime(df.index)  # Convert index to datetime if not already
last_date = df.index[-1]
forecast_dates = pd.date_range(start=last_date + pd.Timedelta(days=1), periods=90)

# Ensure the forecast_dates are in the proper format
forecast_dates = pd.to_datetime(forecast_dates)

# Plotting
plt.figure(figsize=(14, 7))
plt.plot(df.index, df['Harga Beras'], label='Historical Daily Price')  # Ensure df.index is datetime
plt.plot(forecast_dates, predictions.flatten(), label='Forecasted Price', linestyle='--')  # Flatten predictions if necessary
plt.title('Forecast of Harga Beras 90 Days Ahead')
plt.xlabel('Date')
plt.ylabel('Harga Beras')
plt.legend()
plt.xticks(rotation=45)  # Improve date visibility
plt.tight_layout()  # Adjust layout to make room for date labels
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
from math import sqrt

# Assuming actual_values is a numpy array of the actual values for the forecasted period
actual_values = df['Harga Beras'].iloc[-90:]

# Calculate the Mean Absolute Error between the actual and predicted values
mae = mean_absolute_error(actual_values, predictions)

# Calculate the Mean Squared Error between the actual and predicted values
mse = mean_squared_error(actual_values, predictions)

# Calculate the Root Mean Squared Error between the actual and predicted values
rmse = sqrt(mse)

print('Mean Absolute Error:', mae)
print('Mean Squared Error:', mse)
print('Root Mean Squared Error:', rmse)
```
